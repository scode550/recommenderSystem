```python
def pivot_dataframe_in_batches(df, batch_size=100000):
  """Pivots a dataframe in batches.

  Args:
    df: The dataframe to pivot.
    batch_size: The size of the batches to use.

  Returns:
    A pivoted dataframe.
  """

  pivoted_df = pd.DataFrame()
  for i in range(0, len(df), batch_size):
    batch = df[i:i + batch_size]
    pivoted_batch = batch.pivot(index='userId',columns='movieId',values='rating')
    pivoted_df = pd.concat([pivoted_df, pivoted_batch])
  return pivoted_df


Pratings= pivot_dataframe_in_batches(ratings,100000)
print(Pratings)
```
